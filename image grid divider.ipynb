```python
# Importing Image class from PIL module
from PIL import Image
import os

!mkdir input
!mkdir out
 
#im = Image.open(r"/content/seed10468.png")

folder = '/content/input'
listing = os.listdir(folder)
inputfolder = sorted(listing)
```

```python
filecount = 0

for filename in inputfolder:

    #im = "/content/input/a_{}_out.png".format(quarters_iter)
    #im = "/content/input/{}".format(filename)
    im = Image.open(r"/content/input/{}".format(filename))

    #set how  many rows and columns to divide into
    divider = 4
    
    # Size of the image in pixels (size of original image)
    width, height = im.size
    dim1 = width
    dim2 = dim1/divider
    
    count = 0
    for x in range(divider):

      for y in range(divider):

        left = dim2*y
        top = dim2*x
        right = left + dim2
        bottom = top + dim2

        im1 = im.crop((left, top, right, bottom))
        #path = "a_{}.jpg".format(count)
        #im1.save(path)

        save_path = "/content/out"
        im1.save(f"{save_path}/{filecount}_{count}.png")

        count += 1

    filecount += 1
```
